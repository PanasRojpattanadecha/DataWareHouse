* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python
# Check for missing values
missing_values = titanic.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Analyze missing values percentage
total_rows = len(titanic)
missing_percentage = (missing_values / total_rows) * 100
print("\nMissing values percentage in each column:\n", missing_percentage)

# Explore outliers and noisy data (This requires more in-depth analysis)
# For example, you can use box plots or histograms to visualize potential outliers
# Here's a basic example for the 'Age' column:
plt.boxplot(titanic['age'].dropna())
plt.title('Boxplot of Age')
plt.show()

# You can similarly analyze other columns to identify outliers and noisy data
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# ตรวจสอบค่าว่างในแต่ละคอลัมน์
missing_values = titanic.isnull().sum()
print("Missing values per column:\n", missing_values)

# วิธีการจัดการกับค่าว่าง:
# 1. แทนที่ค่าว่างในคอลัมน์ตัวเลข เช่น age ด้วยค่าเฉลี่ย
titanic['age'].fillna(titanic['age'].mean(), inplace=True)

# 2. ลบแถวที่มีค่าว่างในคอลัมน์ที่จำเป็นต้องมีข้อมูล เช่น fare
titanic.dropna(subset=['fare'], inplace=True)

# 3. แทนที่ค่าว่างในคอลัมน์ที่เป็นประเภทข้อความด้วย 'Unknown' หรือค่าที่เหมาะสม
titanic['cabin'].fillna('Unknown', inplace=True)

# ตรวจสอบข้อมูลหลังจัดการค่าว่าง
print(titanic.info())

# Visualize missing values after handling
plt.figure(figsize=(10, 6))
missing_after = titanic.isnull().sum()
missing_after.plot(kind='bar')
plt.title('Missing Values After Handling')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.show()

#แทนที่ค่าว่างในคอลัมน์ตัวเลข (age) ด้วยค่าเฉลี่ย: การใช้ค่าเฉลี่ยเป็นวิธีที่เหมาะสมสำหรับคอลัมน์ตัวเลข เช่น อายุ (age) ซึ่งช่วยรักษาข้อมูลเชิงปริมาณของข้อมูลโดยไม่ลบแถวออก
#ลบแถวที่มีค่าว่างในคอลัมน์ที่จำเป็นต้องมีข้อมูล (fare): การลบแถวที่ขาดข้อมูลค่าโดยสาร (fare) ซึ่งเป็นข้อมูลสำคัญที่ต้องใช้ในการวิเคราะห์
#จัดการค่าว่างในคอลัมน์ประเภทข้อความ (cabin) โดยการเติมค่า 'Unknown': สำหรับคอลัมน์ที่มีข้อความ เช่น หมายเลขห้อง (cabin), การเติมด้วยค่า 'Unknown' ช่วยให้ข้อมูลยังคงอยู่และสามารถวิเคราะห์ได้ต่อไป
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# One-hot encoding for Embarked
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='embarked')
titanic = pd.concat([titanic, embarked_encoded], axis=1)
titanic.drop('embarked', axis=1, inplace=True)
```

```python
# Convert 'Sex' to numerical values
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
```

```python
# Create FamilySize column
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
```

```python
# Create IsAlone column
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8])
```

```python
# i. Min-Max Scaling ตัวแปร Fare และ Age
scaler = MinMaxScaler()

# ตรวจสอบว่าคอลัมน์มีค่าว่างหรือไม่ หากมีค่าว่างให้แทนที่ด้วยค่าเฉลี่ยก่อน
titanic['fare'].fillna(titanic['fare'].mean(), inplace=True)
titanic['age'].fillna(titanic['age'].mean(), inplace=True)

# นำตัวแปร Fare และ Age มาทำ Min-Max Scaling
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])

# ii. สร้างตัวแปรใหม่ Fare_Per_Person โดยการนำ Fare มาหารด้วย FamilySize + 1
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']  # สร้างตัวแปร FamilySize หากยังไม่มี
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)

# iii. สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุ
bins = [0, 12, 18, 50, 100]  # กำหนดช่วงของกลุ่มอายุ
labels = ['Child', 'Teenager', 'Adult', 'Senior']  # กำหนดชื่อกลุ่มอายุ
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=bins, labels=labels, right=False)

# ดูข้อมูลหลังจากการ Normalize และสร้างตัวแปรใหม่
print(titanic[['fare', 'age', 'Fare_Per_Person', 'AgeGroup']].head())
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
# Histograms for Age, Fare
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(titanic['age'], bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(titanic['fare'], bins=20)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Box plots for Age, Fare, and Pclass
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.boxplot(titanic['age'])
plt.title('Boxplot of Age')

plt.subplot(1, 2, 2)
plt.boxplot(titanic['fare'])
plt.title('Boxplot of Fare')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
titanic.boxplot(column='fare', by='pclass')
plt.title('Boxplot of Fare by Pclass')
plt.show()
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
# i. ความสัมพันธ์ระหว่าง survived และ pclass
# การสร้างกราฟ Countplot เพื่อดูการรอดชีวิตในแต่ละชั้นโดยสาร

# Import the seaborn library
!pip install seaborn
import seaborn as sns

plt.figure(figsize=(8,6))
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# ii. ความสัมพันธ์ระหว่าง survived และ sex
# การสร้างกราฟ Countplot เพื่อดูการรอดชีวิตในแต่ละเพศ
plt.figure(figsize=(8,6))
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# iii. ความสัมพันธ์ระหว่าง survived และ age
# การใช้ KDE Plot (Kernel Density Estimate) เพื่อดูการกระจายตัวของอายุสำหรับผู้รอดชีวิตและไม่รอดชีวิต
plt.figure(figsize=(8,6))
sns.kdeplot(titanic[titanic['survived'] == 1]['age'], label='Survived', shade=True)
sns.kdeplot(titanic[titanic['survived'] == 0]['age'], label='Not Survived', shade=True)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()

# iv. ความสัมพันธ์ระหว่างจำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare
# การสร้างกราฟ Scatterplot เพื่อดูความสัมพันธ์ระหว่างจำนวนสมาชิกในครอบครัวที่รอดชีวิตกับค่าโดยสาร
plt.figure(figsize=(8,6))
sns.scatterplot(x='FamilySize', y='fare', hue='survived', data=titanic)
plt.title('Fare vs Family Size by Survival')
plt.xlabel('Family Size')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
# สร้างตารางสรุปข้อมูลโดยแบ่งตาม survived
summary_table = titanic.groupby('survived').agg({
    'sex': ['mean'],  # เพศ (เพศชาย 0, เพศหญิง 1)
    'pclass': ['mean', 'std'],  # ชั้นโดยสาร
    'age': ['mean', 'std'],  # อายุ
    'fare': ['mean', 'std'],  # ค่าโดยสาร
    'FamilySize': ['mean', 'std'],  # ขนาดครอบครัว
    'IsAlone': ['mean'],  # เดินทางคนเดียวหรือไม่
}).reset_index()

# ดูตารางสรุปข้อมูล
print(summary_table)

# สรุปข้อมูลเพศเป็นตัวอักษร
summary_table['sex_mean'] = summary_table['sex']['mean'].apply(lambda x: 'Female' if x > 0.5 else 'Male')

# ลบคอลัมน์ที่ซ้ำซ้อน
#summary_table = summary_table.drop(columns=[('sex', 'mean')], level=0) # The column ('sex', 'mean') was renamed in the line above.
summary_table = summary_table.drop(columns=['sex'], level=0) # Drop the multi-level column 'sex'

# ดูตารางสรุปข้อมูลหลังจากแก้ไข
print(summary_table)
```
